In order to make practice with NN, explore how does the previous linear regression depend on the number of epochs, $N_{\mathrm{epochs}}$, the number of data points $N_{\mathrm{train}}$ and on the noise $\sigma$. Try to improve the previous result operating on these parameters.<br> <br>
The script below generates an initial set of a fixed size of training and testing values, generated according to the function $y(x) = mx + b$, for demonstration purposes.

## <span style="color:green"> Solution </span>


```python
# Prepare initial regression
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import logging
import sys, os
import pathlib
from datetime import datetime
from IPython.display import display, clear_output
from IPython.display import Markdown as md


logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO)
fh = logging.FileHandler('{:%Y-%m-%d}.log'.format(datetime.now()))
formatter = logging.Formatter('%(asctime)s | %(levelname)-8s | %(lineno)04d | %(message)s')
fh.setFormatter(formatter)

logger.addHandler(fh)
from scipy.optimize import curve_fit

def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)

m = 2
b = 1

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 100)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
from keras_visualizer import visualizer 
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects, plot_model

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,))) # 1 input neuron

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.summary()
print()
history = model.fit(x=x_train, y=y_train, 
    batch_size=32, epochs=100,
    shuffle=True, #
    validation_data=(x_valid, y_valid),
    verbose=0
)

score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()

os.makedirs("./models", exist_ok=True)

model.save('./models/model_ex1')
print("Model saved")
```

```python
# Fitting functions
def power_law(x,a,b):
    return a*math.pow(x,b)
power_law_vec = np.vectorize(power_law)

def exp_law(x,a,b):
    return a*math.exp(-b*x)
exp_law_vec = np.vectorize(exp_law)
```

## <span style="color:green"> Dependance on number of epochs </span>
The following section shows the model dependance on the number of epochs for training.
### Method

A for loop has been defined in a range of epochs [e_min,e_max] with a step e_step, and inside each loop the following actions are taken:
- Train the model using training data, and a certain number of epochs
- Get the weights for each neuron
- Evaluate the testing data using the same batch_size as the training one
- Save the accuracy of the model on the testing data

The graph for the dependance of the model testing accuracy on the number of epochs is then graphed.

```python
epochs_min = 10
epochs_max = 200
epochs_step = 5

N_epochs = np.arange(epochs_min,epochs_max,epochs_step)
y_epochs = []
for i,n in enumerate(N_epochs): 
    tmpmodel1 = tf.keras.Sequential()
    tmpmodel1.add(Dense(1, input_shape=(1,))) # 1 input neuron

    # compile the model choosing optimizer, loss and metrics objects
    tmpmodel1.compile(optimizer='sgd', loss='mse', metrics=['mse','accuracy'])
    clear_output(wait=True)
    print(f"Iteration {i}/{len(N_epochs)} --> N epochs = {n}/{epochs_max}")
    history = tmpmodel1.fit(
        x=x_train, y=y_train, 
        batch_size=32, epochs=n,
        shuffle=True, # a good idea is to shuffle input before at each epoch
        validation_data=(x_valid, y_valid),
        verbose=0
    )

    model.get_weights()
    
    score = tmpmodel1.evaluate(x_valid, y_target, batch_size=32, verbose=0)
    # print performance
    print()
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    y_epochs.append(score[1])

print(len(N_epochs),"  ",len(y_epochs))

plt.title("Behaviour of test accuracy with epochs")
plt.xlabel("N° of epochs")
plt.ylabel("Test accuracy")
plt.scatter(N_epochs,y_epochs,label="model_accuracy")
plt.plot(N_epochs,y_epochs,label="model_accuracy")
popt, pcov = curve_fit(power_law_vec, N_epochs, y_epochs,p0=[2,-2])
#plt.plot(N_epochs, power_law_vec(N_epochs, *popt), 'r-',label='power fit: a=%5.3f, b=%5.3f' % tuple(popt))
plt.legend()
plt.grid(True)
plt.show()
```

## <span style="color:green"> Dependance on sample size </span>
In this section, I decided to study the ideal value for the trade-off of training set size and validation set size, exported by a fixed-size dataset, in order to minimize the accuracy of the model. The critical point for the trade-off stricly depends on the type and size of dataset, making it different for each problem. Generally a training set which is too small does not train the model accurately enough, as the error scales with the number of samples. The higher the size of the training set, the more predictive power the model might have, however, after a certain threshold, the estimate might be as accurate as it needs to be, therefore "wasting" a certain amount of dataset. <br>
This means that the theoretical trade-off threshold must be in between the range [0,1].

```python
testsize_percentages = np.arange(0.1,1,0.05)
sample_size = 1000
nepochs = 60
nbatch = 32
ydata = []

for it,perc in enumerate(testsize_percentages):
    clear_output(wait=True)
    tmpmodel = tf.keras.Sequential()
    tmpmodel.add(Dense(1, input_shape=(1,))) # 1 input neuron

    # compile the model choosing optimizer, loss and metrics objects
    tmpmodel.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, int(sample_size*(1-perc))+1)
    x_valid = np.random.uniform(-1, 1, int(sample_size*perc))
    x_valid.sort()
    y_target = m * x_valid + b # ideal (target) linear function
    sigma = 0.0 # noise standard deviation, for the moment it is absent
    y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(m * x_valid + b, sigma)
    w = tmpmodel.get_weights()
    print(f"Iteration {it} of {len(testsize_percentages)-1}")
    print("Initial weights: ",w)
    print(f"Fitting model with parameters:\n- Training data size: {len(x_train)}/{sample_size} --> {(1-perc)*100}% \n- Testing data size: {len(x_valid)}/{sample_size} --> {perc*100}%\n- N° epochs: {nepochs}\n- Batch size: {nbatch}")
    history = tmpmodel.fit(
        x=x_train, y=y_train, 
        batch_size=nbatch, epochs=nepochs,
        shuffle=True, # a good idea is to shuffle input before at each epoch
        validation_data=(x_valid, y_valid),
        verbose=0
    )
    score = tmpmodel.evaluate(x_valid, y_valid, batch_size=32, verbose=0)
    print("Final weights: ",tmpmodel.get_weights())
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    ydata.append(score[1])
    

plt.title("Model accuracy with sample size")
plt.xlabel("Test size")
plt.ylabel("Testing accuracy")
plt.scatter(testsize_percentages,ydata)
plt.plot(testsize_percentages,ydata)
plt.grid(True)
plt.show()
```
