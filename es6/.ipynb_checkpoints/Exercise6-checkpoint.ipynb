## <span style="color:green"> Solution </span>
In order to complete the exercise, the program is based on the ```ising.hpp``` module with methods to compute an Ising model, a ```main.cpp``` file to run the simulation and a ```database.hpp``` library to save results into a database using SQLite.
The main program takes as command-line inputs:
- configfile: the file defining some input variables
- nrestart: the number of restart iteration of the simulation
- dbname: the name of the database where outputs will be stored,
and computes the datablocking of the requested variables.

The simulation automator is defined in the ```run.py``` file, where different temperatures,h-values and sampling methods are defined and eventually fed to the simulation. Each set of parameters defined a unique database that stored values to be plot.

Below, a graph for each variable at fixed parameters is plotted against the number of blocks to show the convergence to the expected value. 

````python
import numpy as np
import sys,os
sys.path.append("./code/")
import utils
from matplotlib import pyplot as plt
from matplotlib import colors as clrs
keys = ("TEMP","BETA","NSPIN","J","H","SAMPLING","NSTEPS","BLOCKS")
variables = ("energy","heat","susc","magn")
samplings = ("metropolis","gibbs")
colors = ("orange","red","green","purple")

def limit(x):
    return float("{:.6f}".format(x))

temps = np.linspace(0.2,3.0,num=20)
J                = 1.0
h                = 0.02
n_spins          = 50

beta             = 1 / temps
th               = np.tanh(beta * J)
thN              = th**n_spins
ch = 1/th

l1 = (np.exp(beta * J) * np.cosh(beta * h) +
    np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) - 2 * np.sinh(2 * beta * J)))
l2 = (np.exp(beta * J) * np.cosh(beta * h) -
    np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) - 2 * np.sinh(2 * beta * J)))
Z = l1**n_spins + l2**n_spins

expected = {
    "energy" : -J * (th + ch * thN) / (1 + thN),
    "heat" : ((beta * J)**2) * (
        ((1 + thN + (n_spins - 1) * (th**2) + (n_spins - 1) * (ch**2) * thN) / (1 + thN)) -
        n_spins *((th + ch * thN) / (1 + thN))**2),
    "susc" : beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN),
    "magn" : 1 / Z * (np.exp(beta * J) * np.sinh(beta * h) * (
        (l1**(n_spins-1)) *
        (1 + np.exp(beta * J) * np.cosh(beta * h) / np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) -
            2 * np.sinh(2 * beta * J))) +
        (l2**(n_spins - 1)) *
        (1 - np.exp(beta * J) * np.cosh(beta * h) / np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) -
            2 * np.sinh(2 * beta * J)))
        )
    )
}
````

In the graph below, a visualization of the final configuration is shown, where the red color means a spin with value -1, while a blue color means a spin value of +1.

````python
spins = np.loadtxt("./final_configs/config.final", unpack=True)
spins = spins.reshape((5,10))

# create discrete colormap
cmap = clrs.ListedColormap(['red', 'blue'])
bounds = [-1,1]
norm = clrs.BoundaryNorm(bounds, cmap.N)


fig, ax = plt.subplots(figsize=(6,4))
plt.suptitle("Final spin configuration: red=down, blue=up")
ax.imshow(spins, cmap=cmap, norm=norm)

# draw gridlines
ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=2)
ax.set_xticks(np.arange(0.5, 10.5, 1));
ax.set_yticks(np.arange(0.5, 5.5, 1));
plt.show()
````

Below, a few graphs show the convergence of the value of the 4 macroscoping variables after datablocking as a function of the number of block, together with the statistical error. The parameters are:
- temperature = 3
- h = 0
- sampling = metropolis
as shown from the string representation of the database reader

````python
fig, ax = plt.subplots(2,2,figsize=(12,10))
reader = utils.DatabaseHandler(temp=3,nspin=50,j=1.0,h=0, sampling="metropolis")
print(reader)
queries = reader.search(False)
colsize = len(variables)/2
plt.suptitle("Blocking of variables for fixed parameters")
for i,v in enumerate(variables):
    x = list(zip(*queries[0].get_variable(v)))
    ind1 = int(i/colsize)
    ind2 = int(i%colsize)
    ax[ind1][ind2].errorbar(x[0][:100],x[1][:100],yerr=x[2][:100],color=colors[i])
    ax[ind1][ind2].set_title(v.capitalize())
    ax[ind1][ind2].set_ylabel(v.capitalize())
    ax[ind1][ind2].set_xlabel("Block")
    ax[ind1][ind2].grid(True)
plt.show()
````

What we are looking for, is the behaviour of each variable as a function of the temperature T plotted by using Metropolis and Gibbs sampling, and compared to the calculated analytical value.

````python

for v in variables:
    fig, ax = plt.subplots(figsize=(10,6))
    y = {
        "metropolis" : [],
        "gibbs" : []
    }
    
    for t in temps:
        for it, s in enumerate(samplings): 
            if v=="magn":
                reader = utils.DatabaseHandler(temp=limit(t),nspin=50,j=1.0,h=0.02,sampling=s)
            else:
                reader = utils.DatabaseHandler(temp=limit(t),nspin=50,j=1.0,h=0,sampling=s)
            queries = reader.search(False)
            values = list(zip(*queries[0].get_variable(v)))
            val = values[1][-1]
            y[s].append(val)
            
    plt.title(v.capitalize())
    plt.plot(temps,y["metropolis"],linewidth=1,label="Metropolis")
    plt.plot(temps,y["gibbs"],linewidth=1,label="Gibbs")
    plt.scatter(temps,y["metropolis"])
    plt.scatter(temps,y["gibbs"])
    plt.plot(temps,expected[v],label="Analytical value")
    plt.ylabel(v.capitalize()+" [LJ]")
    plt.xlabel("Temperature [K]")
    plt.legend()
    plt.grid(True)
    plt.show()
````
