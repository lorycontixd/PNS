## <span style="color:green"> Solution </span>
### <span style="color:red"> The problem </span>
The Travelling Salesman Problem (TSP) is a NP-Hard problem that poses the following question: "Given a list of cities and the distance between each of them, what is the shortest route that visits each city and returns to the starting one?".

### <span style="color:red"> Structure </span>
The simulation for the TSP is made up of numerous files (for more information on the functionalities, read the in-file documentations):
- **genetic.hpp**: contains the class for the implementation of the Genetic Algorithm
- **population.hpp**: contains the population class which represents the population of the Genetic Algorithm
- **individual.hpp**: contains the individual class which represents the individual, or <u>a</u> solution to the considered problem
- **citygen.hpp**: set of functions to generate the cities (index, x-coordinate & y-coordinate) and save them in a file
- **metrics.hpp**: classes representing the metric to use for calculating the distances between cities. The file is made up of an abstract parent class and 2 child classes representing the L1 and L2 metrics:

$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$
$$
L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2
$$
- **mutations.hpp**: various functions for the mutation of individuals
- **random.hpp**: pseudo-random number generator class
- **exceptions.hpp**: custom error system for the Genetic Algorithm. The system is made up of a parent class called BaseError that inherits from the std::exception class. The errors are called through the overridden ```what()``` method.
- **database.hpp**: includes the database class used to save the results of each simulation using the C++ sqlite3 driver.<br>Each simulation produces:
<ol>
    <li> A sequence of genes representing the best individual of the final population. </li>
    <li> The value of fitness for the best individual of each generation. </li>
    <li> The value of fitness for the last generation. </li>
    <li> A set of parameters describing the simulation.</li>
</ol>
<img src="../media/images/es9_db_structure.png" alt="DB Structure" width="242" height="242", class="center">
<br>
- **utility.py**: contains a set of utility functions for the simulation and a class to access and read the sqlite3 database.
- **main.cpp**: contains the execution of the simulation.<br><br>

### <span style="color:red"> Code </span>
The script reads the configurational file from command-line and initializes the Genetic Algorithm. This loads the simulation settings, the database and its settings, it generates the cities and sets an initial population of individuals. Until the value of fitness converges, the simulation obeys the following steps:
<ol>
    <li>Selects two parents at random from the population individuals.</li>
    <li>Mutates both parents with a probability of 10%.</li>
    <li>Performs a crossover between the two mutates parents to produce two offsprings.</li>
    <li>Adds the offsprings to the new population. These steps are repeated until the number of individuals of the new population matches the number of the old population.</li>
</ol>
Lastly, the results are saved in the database and the simulation is terminated.

In this notebook, a python script has been written to iterate over a bunch of parameters to be fed to the simulation, in order to compare the speed of convergence as a function of these parameters.

````python
"""
The classes and functions used in this notebook are defined in utility.py to keep this notebook clean.
In order to visualize the final outcome and best trip of the simulation, a line collection from matplotlib has been used.
"""
from utility import *
import numpy as np
import configparser as cp
import sys, os
import subprocess
import pylab as pl
from matplotlib import pyplot as plt
from numba import jit
from matplotlib import collections  as mc
````

# <span style="color:green"> Circle </span>
Below is an graph with the positions of the cities generated on a circumference.<br>
The cities have been defined by generating a random number between 0 and $2\pi$, and then selecting the x and y coordinates by applying the sine and cosine functions respectively, for each city iteration.

````python
LAYOUT = "circle"
print("\n---------  GENERATED CITIES  ---------\n")
cityi, cityx, cityy = np.loadtxt(f"./outputs/{LAYOUT}/cityposition.dat",unpack=True)
s = [4 for k in range(len(cityi))]
fig, ax = plt.subplots(figsize=(14,6))
ax.scatter(cityx, cityy,color="blue",s=s)
for index, txt in enumerate(cityi):
    ax.annotate(int(txt), (cityx[index], cityy[index]))
````

In the cell below, different parameters are explored and fed to the Genetic Algorithm simulation.<br>
Both types of metrics are used, a few values of the circumference radius, a bunch of values for the number of individuals to be explored and various numbers for the final generation value.
For each simulation, a temporary configuration file (tempfile.ini) is created with a dedicated function and the simulation is called with a subcommand.

````python
output_level = 1 # 0=outputs nothing
nindividuals = [100,200,500,1000]
npopulations = [100,50,25,10,5]
metrics = ["L1","L2"]
boxes = [1,5,50]
total = len(npopulations)*len(metrics)*len(boxes)*len(nindividuals)
j=1
clear_previous_databases(LAYOUT)
percentages = [int(i*total/100) for i in np.arange(0,110,10)]
print("-- Starting a total of %i simulations --" % total)

for npop in npopulations:
    for nind in nindividuals:
        for met in (metrics):
            for box in boxes:
                if output_level>0:
                    if j in percentages:
                        print("Executing simulation ",j,"/",total)
                create_temp_ini(LAYOUT,32,box,met,nind,npop)
                output = os.popen("make && ./main tempfile.ini").read()
                if output_level>1:
                    print(output)
                j+=1
print("-- Simulation completed for circle --")
````

## Maximum number of generations
The following section aims to study the convergence of the fitness value as a function of the number of generations allowed. It is expected that a number of generations which is too small does not allow for full convergence, while a number of generations which is too high takes unnecessary computational time, because the algorithm has already converged.

````python
s = DatabaseReader(max_pop=100,layout=LAYOUT,metric="L2",box=1,max_ind=500)
print(s)
fig, ax = plt.subplots(figsize=(12,5))
ax.set_title(f"Fittest: Fitness over generation - max={s.get_setting_value('max_pop')} ")
ax.set_xlabel("Generation")
ax.set_ylabel("Fitness")
ax.grid(True)
ax.plot(s.get_fpi(),linewidth=2)
plt.show()
print("\nFinal fitness: ",s.get_fitness(s.get_setting()),"\n")
````

````python
s = DatabaseReader(max_pop=10,layout=LAYOUT,metric="L2",box=1,max_ind=500)
print(s)
fig, ax = plt.subplots(figsize=(12,5))
ax.set_title(f"Fittest: Fitness over generation - max={s.get_setting_value('max_pop')} ")
ax.set_xlabel("Generation")
ax.set_ylabel("Fitness")
ax.grid(True)
ax.plot(s.get_fpi(),linewidth=2)
plt.show()
print("\nFinal fitness: ",s.get_fitness(s.get_setting()),"\n")
````

## Number of Individuals
The number of individuals in a population represents how many solutions are explored for each generation. The expectation is that a higher number of individuals allows for a faster convergence, because the optimal solution to the problem is found at earlier stages of the algorithm

````python
fig, ax = plt.subplots(figsize=(12,5))
ax.set_title("Fitness comparison for number of max individuals")
ax.set_xlabel("Generation")
ax.set_ylabel("Fitness")
for inds in nindividuals:
    reader = DatabaseReader(max_pop=50,max_ind=inds,layout=LAYOUT,metric="L2",box=1)
    ax.plot(reader.get_fpi(),label=f"N ind: {inds}")
ax.grid(True)
plt.legend()
plt.show()
````

## Metrics

````python
l2 = DatabaseReader(max_pop=100,layout=LAYOUT,metric="L2",box=1)
l1 = DatabaseReader(max_pop=100,layout=LAYOUT,metric="L1",box=1)

fig, ax = plt.subplots(figsize=(12,5))
ax.set_title("Fitness comparison for metrics")
ax.set_xlabel("Generation")
ax.set_ylabel("Fitness")
ax.plot(l1.get_fpi(),label="L1")
ax.plot(l2.get_fpi(),label="L2")
ax.grid(True)
plt.legend()
plt.show()
````

Below is a figure representing the best path of the traveller between all cities.

````python
best = DatabaseReader(max_ind=500,max_pop=100,box=1,layout=LAYOUT,metric="L2")
print(f"--->  SHOWING BEST PATH\n")
genes = best.get_best()
print(genes)
fig, ax = plt.subplots(figsize=(12,6))
s=[2 for i in range(len(genes))]
ax.scatter(cityx,cityy,s=s)
for index, txt in enumerate(cityi):
    ax.annotate(int(txt), (cityx[index], cityy[index]))
lines = []
c = []
for k in range(len(genes)-1):
    lines.append([(cityx[genes[k]],cityy[genes[k]]),(cityx[genes[k+1]],cityy[genes[k+1]])])
    c.append((1,0,0,1))
lc = mc.LineCollection(lines, colors=c, linewidths=2)
ax.add_collection(lc)
ax.autoscale()
ax.margins(0.1)
plt.show()
````
